```python
from pymongo import MongoClient
import pickle
from bs4 import BeautifulSoup
```

```python
db_name = 'airbnb'
coll_name = 'search'

client = MongoClient()
db = client[db_name]
coll = db[coll_name]
```

```python
x = coll.find_one()
```

```python
x
```

```python
r = pickle.loads(x['pickle'])
```

```python
soup = BeautifulSoup(r.content)
```

```python
air_sr = soup.find_all("div", {"class": "listing"})[0]
```

```python
#data
air_sr['data-id']
air_sr['data-lat']
air_sr['data-lng']
air_sr['data-name']
air_sr['data-price']
int(air_sr['data-price'].replace('</sup>','').replace('<sup>','').replace('$',''))    # note this isn't great
air_sr['data-url']
air_sr['data-user']
```

```python
listing_ids = []

for x in coll.find():
    r = pickle.loads(x['pickle'])
    soup = BeautifulSoup(r.content)
    for listing in soup.find_all("div", {"class": "listing"}):
        listing_id = listing['data-id']
        listing_ids.append(listing_id)
```

```python
print len(listing_ids)    # list of all ids I've found
print len(set(listing_ids))    # unique ids
```

```python
sorted(set(listing_ids), key=lambda x: int(x))
```
