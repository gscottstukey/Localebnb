```python
listing_ids = []

for x in coll.find():
    r = pickle.loads(x['pickle'])
    soup = BeautifulSoup(r.content)
    for listing in soup.find_all("div", {"class": "listing"}):
        listing_id = listing['data-id']
        listing_ids.append(listing_id)
```

```python
print len(listing_ids)    # list of all ids I've found
print len(set(listing_ids))    # unique ids
```

```python
sorted(set(listing_ids), key=lambda x: int(x))
```

#Neighborhoods

```python
import pickle
```

```python
db_name = 'airbnb'
coll_name = 'neighborhoods'

client = MongoClient()
db = client[db_name]
coll = db[coll_name]
```

```python
x = coll.find_one()
```

```python
for x in coll.find():
    if 'error' in x:
        print "ERROR: %s" % x['_id']
```

```python
hood = coll.find_one()
```

```python
r = pickle.loads(hood['pickle'])
```

```python
soup = BeautifulSoup(r.content)
```

```python
soup.find('div', {'class':'center description'}).get_text().strip()    # Hood Headline
soup.find('meta', {'name':'description'})['content']    # Hood Description

traits = []
for trait in soup.find('ul', {'class':'traits'}).find_all('span'):
    traits.append(trait.get_text())
traits    # traits

tags = []
for tag in soup.find_all('div', {'class':'neighborhood-tag'}):
    tags.append(tag.get_text().strip())
tags    # neighborhood tags

similar_hoods = []
for similar_hood in soup.find('ul', {'class':'trait-neighborhoods neighborhoods'}).find_all('li'):
    similar_hoods.append(similar_hood['data-neighborhood-permalink'])
similar_hoods    # similar hoods

soup.find('div', {'class':'caption bar'}).find_all('li')    # exploratory
soup.find('div', {'class':'caption bar'}).find_all('li')[0].strong.get_text()    # Public Trans
soup.find('div', {'class':'caption bar'}).find_all('li')[1].strong.get_text()    # Having a Car

soup.find('div', {'id':'inner-map'})['data-bbox']   # # map_data_bbox
soup.find('div', {'id':'inner-map'})['data-x']    # map_data_x
soup.find('div', {'id':'inner-map'})['data-y']    # map_data_y
```
